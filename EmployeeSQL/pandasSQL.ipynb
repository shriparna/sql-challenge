```python
import pandas as pd
```

```python
dept_df = pd.read_csv("../data/departments.csv")
dept_df
```

```python
dept_emp_df = pd.read_csv("../data/dept_emp.csv")
dept_emp_df.head()
```

```python
dept_mgr_df = pd.read_csv("../data/dept_manager.csv")
dept_mgr_df.head()
```

```python
emp_df = pd.read_csv("../data/employees.csv")
emp_df.head()
```

```python
sal_df = pd.read_csv("../data/salaries.csv")
sal_df.head()
```

```python
tit_df = pd.read_csv("../data/titles.csv")
tit_df
```

```python
# 1. List the employee number, last name, first name, sex, and salary of each employee.
emp_sal_merged = emp_df.merge(sal_df, how="inner", on="emp_no")
emp_salary_merged = emp_sal_merged.loc[:, ["emp_no", "last_name", "first_name", "sex"]]
emp_salary_merged.head(20)
```

```python
# 2. List the first name, last name, and hire date for the employees who were hired in 1986.
condition_1986= emp_df["hire_date"].str.split('/', expand=True)[2]=="1986"
#condition_1986
emp_1086_df = emp_df.loc[condition_1986,["emp_no", "last_name", "first_name", "hire_date"]] 
emp_1086_df
```

```python
# 3. List the manager of each department along with their department number, department name, employee number, last name, and first name.
dept_dept_mgr_merged_df = dept_df.merge(dept_mgr_df, how="inner", on="dept_no")
dept_dept_mgr_emp_merged_df = dept_dept_mgr_merged_df.merge(emp_df, how="inner", on="emp_no")
dept_dept_mgr_emp_merged_df = dept_dept_mgr_emp_merged_df.loc[:,["dept_no", "dept_name", "emp_no", "last_name", "first_name"]]
dept_dept_mgr_emp_merged_df
```

```python
# 4. List the department number for each employee along with that employee’s employee number, last name, first name, and department name.
dept_dept_emp_merged_df = dept_df.merge(dept_emp_df, how="inner", on="dept_no")
dept_dept_emp_emp_merged_df = dept_dept_emp_merged_df.merge(emp_df, how="inner", on="emp_no")
dept_dept_emp_emp_merged_df = dept_dept_emp_emp_merged_df.loc[:,["emp_no", "last_name", "first_name", "dept_name"]]
dept_dept_emp_emp_merged_df
```

```python
# 5. List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.
condition1 = emp_df["first_name"]=="Hercules"
condition2 = emp_df["last_name"].str.startswith("B")
condition = condition1 & condition2
condition
Hercules_df = emp_df.loc[condition,["first_name", "last_name", "sex"]]
Hercules_df
```

```python
# 6. List each employee in the Sales department, including their employee number, last name, and first name.
sales_dept_df = dept_df.loc[dept_df["dept_name"]=="Sales",:]
sales_dept_emp_df = sales_dept_df.merge(dept_emp_df, how="inner", on="dept_no")
sales_dept_emp_merge_df = sales_dept_emp_df.merge(emp_df, how="inner", on="emp_no")
sales_dept_emp_merge_df = sales_dept_emp_merge_df.loc[:,["emp_no", "last_name", "first_name", "dept_name"]]
sales_dept_emp_merge_df
```

```python
# 7. List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.
sales_dev_dept_df = dept_df.loc[(dept_df["dept_name"]=="Sales") | (dept_df["dept_name"]=="Development"),:]
sales_dev_dept_emp_df = sales_dev_dept_df.merge(dept_emp_df, how="inner", on="dept_no")
sales_dev_dept_emp_merge_df = sales_dev_dept_emp_df.merge(emp_df, how="inner", on="emp_no")
sales_dev_dept_emp_merge_df
sales_dev_dept_emp_merge_df = sales_dev_dept_emp_merge_df.loc[:,["emp_no", "last_name", "first_name", "dept_name"]]
sales_dev_dept_emp_merge_df
```

```python
# 8. List the frequency counts, in descending order, of all the employee last names (that is, how many employees share each last name).
emp_last_name_count_df = emp_df["last_name"].value_counts()
emp_last_name_count_df
```
